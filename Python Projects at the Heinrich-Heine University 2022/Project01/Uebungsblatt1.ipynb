## Aufgabe 1 (2 Punkte)
Schreiben Sie ein Programm, das **alle** Wörter des Gedichts "Für Einen" ausgibt, eins pro Zeile. Die Satzzeichen am Ende der Wörter sollen (wie in der Vorlesung) nicht entfernt werden. Das Gedicht befindet sich in der Datei ``kaleko-fuer-einen.txt``

```python
f = open("kaleko-fuer-einen.txt", encoding="utf-8")
words = f.read()
f.close()

for word in words.split():
    print(word)
```

## Aufgabe 2 (4 Punkte)
Schreiben Sie ein Programm, das alle **kleingeschriebenen** Wörter des Gedichts "Für Einen" ausgibt, eins pro Zeile. Die Satzzeichen am Ende der Wörter sollen (wie in der Vorlesung) nicht entfernt werden.


```python
f = open("kaleko-fuer-einen.txt", encoding="utf-8")
words = f.read()
f.close()

for word in words.split():
    if word.islower():
        print(word)
```

## Aufgabe 3 (4 Punkte)
Schreiben Sie ein Programm, das alle englischen Wörter aus dem Wörterbuch (mit Dateinamen ``american-english.txt``) zeilenweise ausgibt, die Kennzeichen in **Düsseldorf** sein können.

```python
f = open("american-english.txt", encoding="utf-8")
words = f.read()
f.close()

for word in words.split():
    if word.startswith("d") and 2 <= len(word) <= 3:
        print(word)
```
