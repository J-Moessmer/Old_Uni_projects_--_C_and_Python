**Hinweise:** Verwenden Sie ein Dictionary, um Braille-Zeichen auf lateinische Zeichen abzubilden. Gehen Sie bei der Übersetzung zeichenweise vor. Kopieren Sie die Braille-Zeichen aus der Aufgabenstellung, um keinen Fehler zu machen.

````python
f = open("braille.txt", encoding="utf-8")
text = f.read()
f.close()


char_to_replace = {'⠁': 'a',
'⠃': 'b',
'⠉': 'c',
'⠙': 'd',
'⠑': 'e',
'⠋': 'f',
'⠛': 'g',
'⠓': 'h',
'⠊': 'i',
'⠚': 'j',
'⠅': 'k',
'⠇': 'l',
'⠍': 'm',
'⠝': 'n',
'⠕': 'o',
'⠏': 'p',
'⠟': 'q',
'⠗': 'r',
'⠎': 's',
'⠞': 't',
'⠥': 'u',
'⠧': 'v',
'⠺': 'w',
'⠭': 'x',
'⠽': 'y',
'⠵': 'z',
'⠜': 'ä',
'⠪': 'ö',
'⠳': 'ü',
'⠮': 'ß',
'⠾': 'st',
'⠡': 'au',
'⠣': 'eu',
'⠩': 'ei',
'⠌': 'äu',
'⠬': 'ie',
'⠹': 'ch',
'⠱': 'sch',
'⠂': ',',
'⠄': '.',
'⠠': "'",
'⠒': ':',
'⠆': ';',
'⠢': '?',
'⠖': '!'}





for key, value in char_to_replace.items():
    text = text.replace(key, value)
print(text)
````

```

e 187815
n 118688
i 88133
r 78522
...

– 2
‚ 1
```

Das heißt, dass das Leerzeichen 241746 mal vorkommt in der Datei, der Buchstabe ``"e"`` 187815 mal, und das Zeichen ``"*"`` acht mal.

````python
f = open("grimm.txt", encoding="utf-8")
sentence = f.read()
f.close()
#using dict comprehensions
#sentence = "This is a long sentence that I have just written up for an example!"
dictionary = {char:  sentence.count(char) for char in set(sentence)}

#sorted(dictionary)
#dictionary = sorted(dictionary.items(), key = lambda kv:(kv[0], kv[1]))
#print(sorted(dictionary.items(), key=lambda x: x[1], reverse=True))

del dictionary["\n"]
for key, value in sorted(dictionary.items(), key=lambda x: x[1], reverse=True):
    
    print(key, value)



````
