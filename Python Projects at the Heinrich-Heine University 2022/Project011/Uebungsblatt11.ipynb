## Aufgabe 2 (4 Punkte)

Die Datei ``urls.txt`` enthält viele URLs, eine pro Zeile. Schreiben Sie ein Programm, das die Datei einliest und nur den Host jeder URL (zeilenweise) ausgibt. Verwenden sie die Funktion [``urlparse``](https://docs.python.org/3/library/urllib.parse.html#urllib.parse.urlparse) aus dem ``urllib.parse``-Modul, um den Host einer URL zu berechnen. Die ersten Zeilen der Ausgabe sind:

```
lt.wikipedia.org
lv.wikipedia.org
ml.wikipedia.org
ms.wikipedia.org
...
```
