## Aufgabe 1 (2 Punkte)
Der Cartoonist Tom Gauld schlägt eine Methode vor, um Mitglieder für exzentrische Familien in Romanen zufällig zu erzeugen: https://twitter.com/tomgauld/status/861895772975005696

Schreiben Sie ein Programm, das **alle** möglichen so erzeugten Familienmitglieder in alphabetischer Reihenfolge und zeilenweise ausgibt. Die Attribute und Personen des Cartoons sind:

```
attributes = "criminal evil genius imposter kindhearted lazy long-lost millionaire talented"
persons = "baby butler cousin dog grandma guardian nurse sibling stepmother twins uncle"
```

**Hinweis:** Dieses Programm lässt sich mit zwei For-Schleifen lösen, die alphabetische Sortierung ergibt sich automatisch, weil die Attribut- und Personenlisten schon alphabetisch angeordnet sind. Das Attribut kommt *vor* der Person (also etwa ``kindhearted guardian``).

````python
for x in ["criminal", "evil", "genius", "imposter", "kindhearted", "lazy", "long-lost", "millionaire", "talented"]:
    for y in ["baby", "butler", "cousin", "dog", "grandma", "guardian", "nurse", "sibling", "stepmother", "twins", "uncle"]:
        print(x,y)
````

## Aufgabe 2 (4 Punkte)

Erweitern Sie das Wördle-Programm aus der Vorlesung wie folgt:

Das Programm soll richtig mit Umlauten und ß umgehen, dass heißt Ä, Ö, Ü, ß sollen durch AE, OE, UE, SS ersetzt werden. Hierfür verwenden Sie die ``.replace``-Funktion von Strings.

Geben Sie mit dem veränderten Programm die Wörter aus, die auf das Muster `"?P???"` passen!

````python

f = open("german.txt", encoding="utf-8")
words = f.read()
f.close()

pattern = "?P???"

for word in words.split():
    word = word.upper()
    word = word.replace("Ä", "AE")
    word = word.replace("Ü", "UE")
    word = word.replace("Ö", "OE")
    word = word.replace("ß", "SS")
    if len(word) == 5:
        ok = True
        for position in range(len(pattern)):
            searchchar = pattern[position]
            wordchar = word[position]
            if searchchar == wordchar or searchchar == "?":
                pass
            else:
                ok = False
        if ok:
            print(word)
````
